```python

def calc(expression):
    
    number1, number2, operation = parse_expression(expression)
    
    
    fraction1 = convert_to_fraction(number1)
    fraction2 = convert_to_fraction(number2)
    
    if operation == 'делить на':
        
        result = fraction1 / fraction2
        result_fraction = convert_to_fraction(result)
        
        # Проверяем, если результат дробный
        if result_fraction != int(result_fraction):
            # Округляем до тысячных
            rounded_result = round(result_fraction, 3)
            
            # Преобразуем результат в текст
            result_text = convert_to_text(rounded_result)
        else:
            # Преобразуем результат в текст
            result_text = convert_to_text(result_fraction)
        
        return result_text
    
    elif operation == 'остаток от деления на':
        # Вычисляем остаток от деления и приводим его к дроби
        remainder = fraction1 % fraction2
        remainder_fraction = convert_to_fraction(remainder)
        
        # Проверяем, если остаток дробный
        if remainder_fraction != int(remainder_fraction):
            # Округляем до тысячных
            rounded_remainder = round(remainder_fraction, 3)
            
            # Преобразуем остаток в текст
            remainder_text = convert_to_text(rounded_remainder)
        else:
            # Преобразуем остаток в текст
            remainder_text = convert_to_text(remainder_fraction)
        
        return remainder_text

def parse_expression(expression):
    # Разбиваем выражение на числа и операцию
    parts = expression.split()
    number1 = parts[0] + ' ' + parts[1] + ' ' + parts[2]
    number2 = parts[-1]
    operation = ' '.join(parts[3:-2])
    return number1, number2, operation

def convert_to_fraction(number):
    # Преобразуем число в десятичную дробь
    parts = number.split(' и ')
    integer_part = int(parts[0])
    
    if len(parts) > 1:
        decimal_part = sum([1/int(x) for x in parts[1].split()])
    else:
        decimal_part = 0
    
    return integer_part + decimal_part

def convert_to_text(fraction):
    # Преобразуем десятичную дробь в текст
    integer_part = int(fraction)
    decimal_part = round(fraction - integer_part, 3)
    
    if decimal_part == 0:
        return str(integer_part)
    else:
        return f"{integer_part} и {decimal_part} десятых"

# Пример использования
result = calc("сорок один и тридцать одна сотая разделить на семнадцать")
print(result)  # Выводит "два и сорок три сотых"
```
